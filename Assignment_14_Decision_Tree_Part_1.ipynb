```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier,plot_tree
from sklearn.metrics import classification_report,confusion_matrix
```

```python
df=pd.read_csv("C:/Users/pcc/Downloads/Company_Data.csv")
df.head()
```

```python
lb=LabelEncoder()
df[["ShelveLoc","Urban","US"]]=df[["ShelveLoc","Urban","US"]].apply(lb.fit_transform)
df
```

```python
high_sale=[]
for i in df["Sales"]:
    if i>8:
        high_sale.append("High")
    else:
        high_sale.append("Low")
        
```

```python
high_sale
```

```python
df["Sales_high_low"]=pd.DataFrame(high_sale)
df
```

```python
df
```

```python
df.drop("Sales",axis=1,inplace=True)
```

```python
df
```

```python
df["Sales_high_low"].value_counts()
```

```python
df["Sales_high_low"]=lb.fit_transform(df["Sales_high_low"])
```

```python
df
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
x,y
```

```python
df["Sales_high_low"].unique()
```

```python
sns.pairplot(df,hue="Sales_high_low")
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

## Build Desicion Tree Algorithm using Entropy

```python
model=DecisionTreeClassifier(criterion="entropy",max_depth=5)
model.fit(x_train,y_train)
```

```python
plot_tree(model)
plt.show()
```

```python
df.columns
```

```python
fn=['CompPrice', 'Income', 'Advertising', 'Population', 'Price',
       'ShelveLoc', 'Age', 'Education', 'Urban', 'US']
cn=["High","Low"]

plt.figure(figsize=(17,13))
plot_tree(model,max_depth=6,feature_names=fn,class_names=cn,filled=True)
plt.show()
```

```python
pred=model.predict(x_test)
pred
```

```python
pd.Series(pred).value_counts()
```

```python
pd.DataFrame({"Acutual":y_test,
             "Predictd":pred})
```

```python
confusion_matrix(y_test,pred)
```

```python
pred==y_test
```

```python
np.mean(pred==y_test)
```

## Building Desicion Tree Classifier (CART) using Gini Criteria

```python
model_gini=DecisionTreeClassifier(criterion="gini",max_depth=6)
model_gini
```

```python
model_gini.fit(x_test,y_test)
```

```python
plot_tree(model_gini)
plt.show()
```

```python
plt.figure(figsize=(15,8))
plot_tree(model_gini,class_names=cn,
         feature_names=fn,max_depth=6,
         filled=True)
plt.show()
```

```python
pred_1=model_gini.predict(x_test)
```

```python
pd.DataFrame({"Actual":y_test,
             "predicted":pred_1})
```

```python
np.mean(pred_1==y_test)
```

```python
confusion_matrix(y_test,pred_1)
```
