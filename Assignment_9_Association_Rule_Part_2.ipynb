```python
# read file
df=pd.read_csv("my_movies.csv")
df.head()
```

```python
# check unique entries
df["V1"].unique(),df["V2"].unique(),df["V3"].unique(),df["V4"].unique(),df["V5"].unique()
```

```python
#create new dataframe
new_df=df.iloc[:,5:]
```

```python
new_df
```

```python
new_df.shape
```

Ariori Algorithm

min_support=0.1,min_threshold=0.5

```python
frequent_items=apriori(new_df,min_support=0.1,use_colnames=True)
frequent_items
```

```python
rule=association_rules(frequent_items,metric="lift",min_threshold=0.5)
rule
```

```python
# display rule>1 values
rule[rule["lift"]>1]
```

```python
# plot histogram
rule[["support","confidence"]].hist()
```

```python
# histogram
rule[["support","confidence","lift"]].hist()
```

```python
# scatter plot
plt.scatter(rule['support'], rule['confidence'])
plt.show()
```
